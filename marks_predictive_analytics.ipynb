```python
df_raw = df.copy()
```

```python
# ===== 1) Create two separate copies
df_full = df_raw.copy()       # will be preprocessed for the "all-features" model
df_selected = df_raw.copy()   # will be preprocessed for the " selected-features" model
```

## Approach 1: using all features for prediction except for student_id  (df_full)

```python
# ===== 2) Common small fix: fill parental education NA in both (if any)
df_full['parental_education_level'] = df_full['parental_education_level'].fillna(
    df_full['parental_education_level'].mode()[0]
)
df_selected['parental_education_level'] = df_selected['parental_education_level'].fillna(
    df_selected['parental_education_level'].mode()[0]
)
```

```python
# ===== 3) Preprocessing for df_full (ALL FEATURES pipeline)
# 3a) Binary mappings
df_full['part_time_job'] = df_full['part_time_job'].map({'No':0, 'Yes':1})
df_full['extracurricular_participation'] = df_full['extracurricular_participation'].map({'No':0, 'Yes':1})
```

```python
# 3b) Ordinal mappings
diet_map = {'Poor':0, 'Fair':1, 'Good':2}
edu_map  = {'High School':0, 'Bachelor':1, 'Master':2}
internet_map = {'Poor':0, 'Average':1, 'Good':2}

df_full['diet_quality'] = df_full['diet_quality'].map(diet_map)
df_full['parental_education_level'] = df_full['parental_education_level'].map(edu_map)
df_full['internet_quality'] = df_full['internet_quality'].map(internet_map)
```

```python
# 3c) One-hot encode nominal columns (keep all gender dummies explicitly)
df_full = pd.get_dummies(df_full, columns=['gender'], drop_first=False)
```

```python
# 3d) Drop identifier column (not a feature)
df_full = df_full.drop(columns=['student_id'])
```

```python
# 3e) Final check for df_full
print("df_full shape:", df_full.shape)
display(df_full.info())
display(df_full.head())
```

```python
X_full = df_full.drop('exam_score', axis=1)
y_full = df_full['exam_score']
```

```python
X_full.shape, y_full.shape
```

```python
X_full.head()
```

```python
y_full.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_full, y_full, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Decision Tree": DecisionTreeRegressor(random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # compute RMSE using sqrt of MSE to support sklearn versions
    train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

    results[name] = {
        "Train_R2": r2_score(y_train, y_pred_train),
        "Test_R2": r2_score(y_test, y_pred_test),
        "Train_RMSE": train_rmse,
        "Test_RMSE": test_rmse,
    }

results
```

| Model                 | Train R² | Test R²   | Train RMSE | Test RMSE | Interpretation                                                              |
| --------------------- | -------- | --------- | ---------- | --------- | --------------------------------------------------------------------------- |
| **Linear Regression** | 0.901    | 0.898     | ~5.36      | ~5.09     | Very good fit, low bias, low variance                                       |
| **Ridge Regression**  | 0.901    | 0.898     | ~5.36      | ~5.09     | Almost identical to Linear → dataset does not have multicollinearity issues |
| **Decision Tree**     | **1.00** | **0.645** | 0.00       | ~9.53     | **Severe overfitting** (classic behavior)                                   |


```python
import pandas as pd

results_df = pd.DataFrame(results).T
results_df
```

```python
models={
  "LinearRegression":{
    "model":LinearRegression(),
    "params":{}
},
  
  "DecisionTree":{
    "model":DecisionTreeRegressor(),
  "params": {
    'max_depth': [None,3, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
  }
  },
  
  "RandomForest":{
    "model":RandomForestRegressor(),
    "params":{
      'n_estimators': [50,100, 200, 300],
      'max_depth': [None, 5, 10, 20],
      'min_samples_split': [2, 5, 10],
      'min_samples_leaf': [1, 2, 4]
    }
  },
  
  "SVR": {
        "model": Pipeline([
            ("scaler", StandardScaler()),
            ("svr", SVR())
        ]),
        "params": {
            # param names must include the pipeline step prefix: 'svr__'
            "svr__C": [0.1, 1, 10],
            "svr__epsilon": [0.1, 0.5],
            "svr__kernel": ["rbf"]    # keep to rbf for starters
        }
    }
  
}
  
  
```

```python
best_models=[]
```

```python
for name,config in models.items():
    print(f"Training model: {name}")

    grid=GridSearchCV(config['model'],config['params'],cv=5,scoring=['r2','neg_mean_squared_error'],verbose=1,refit='r2',n_jobs=-1)
    grid.fit(X_train,y_train)
    y_pred=grid.predict(X_test)
    
    rmse=np.sqrt(mean_squared_error(y_test,y_pred))
    r2=r2_score(y_test,y_pred)
    
    best_models.append({
        "Model":name,
        "Best_Params":grid.best_params_,
        "Test_R2":r2,
        "Test_RMSE":rmse
    })
```
